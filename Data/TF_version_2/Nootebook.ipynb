## Search of Transcription Factors domains in Cuatro Cienegas and Abyssal genomes

Actual accesion number of TF domains were recovered from Pfam-A.hmm file using TF_domains file

```python
%%bash
cut TF_domains -f1 | while read line; do grep $line ../GENOMES/Pfam-A.hmm | cut -d ' ' -f 4; done > TF_domains_v2
```

Each Pfam profile in TF_domains_v2 were recovered from file Pfam-A.hmm and used to look for in each genome with function hmmsearch

```python
%%bash
ls genomic/*faa | while read line
    do
    cat TF_domains_v2 | hmmfetch -f ../GENOMES/Pfam-A.hmm - |hmmsearch - $line > $line.out
    done
```

To recover the protein sequence in each genome script PFAM-parsel.pl were used

```python
%%bash
ls genomic/*.out | while read line; do perl ../../Sources/pFAM-parser.pl $line; done
```

Aminoacid sequence were recovered

```python
%%bash
ls genomic/*out-PF* | while read list
    do
    cut $list -f2- | tail -n +2 | while read id
        do
        echo '>'$id >> $list.fasta
        grep -A 1000 -w -h $id genomic/*.faa | sed -n -e '1,/>/ {/>/ !{'p''}} >>$list.fasta
        done
    done
```

Sequences of same protein family were grouped in a same file

```python
%%bash
cat TF_domains_v2 | while read line
    do echo $line
    cat genomic/*$line*fasta > genomic/$line'_all.fasta'
    done
```

## Search of Transcription Factors domains in MIBiG clusters

Proteins with TF domain in MIBiG clusters were search

```python
%%bash
ls MIBIG/mibig_prot_seqs_ids.fasta |while read line
    do
    cat TF_domains_v2 | hmmfetch -f ../GENOMES/Pfam-A.hmm - |hmmsearch - $line > $line.out
    done
```

ID of proteins were recovered with script pFAM-parser.pl

```python
%%bash
ls MIBIG/*out* | while read line
    do perl ../../Sources/pFAM-parser.pl $line
    done
```

IDs were corrected to match with actual IDs

```python
%%bash
perl -p -i -e 's/\n/_/g' MIBIG/*PF*
```

```python
%%bash
perl -p -i -e 's/_BGC0/\nBGC0/g' MIBIG/*PF*
```

```python
%%bash
perl -p -i -e 's/_$/\n/' MIBIG/*PF*
```

```python
%%bash
sed -i 1d MIBIG/*PF*
```

Aminoacid sequences were recovered

```python
%%bash
ls MIBIG/*out-PF* | while read list
    do
    cat $list | while read id
        do
        echo '>'$id >> $list.fasta
        grep -A 1000 -w -h $id MIBIG_v2/mibig_prot_seqs_ids.fasta | sed -n -e '1,/>/ {/>/ !{'p''}} >> $list.fasta
    done
done
```

Sequences of same protein family were grouped in a same file

```python
%%bash
cat TF_domains_v2 | while read line
    do
    echo $line
    cat MIBIG/*$line*fasta > MIBIG/$line'_all.fasta'
    done
```

Sequences of genomic and MIBiG were combined

```python
%%bash
cat TF_domains_v2 | while read line
    do
    echo $line
    cat 'genomic/'$line'_all.fasta' 'MIBIG/'$line'_all.fasta' > TF_genomic/$line'_genomic.fasta'
    done
```

## Clusteric TA sequences

IDs belonging to Biosynthesis Gene Clusters (BGC), "clusteric", were filtered using clusteric genes metadata

```python
%%bash
ls genomic/PF* | while read file
    do
    grep '>' $file | cut -d '>' -f 2 | while read id
        do
        grep -w -h -s -o $id TF_clusteric/metadata.txt >> $file'_clusteric.id'
        done
    mv $file'_clusteric.id' TF_clusteric/
    done
```

Aminoacid sequences from clusteic IDs were recovered

```python
%%bash
ls TF_clusteric/*id | while read line
    do
    cat $line | while read id
        do
        echo '>'$id >> $line.fasta
        grep -A 1000 -w -h $id genomic/*all.fasta | sed -n -e '1,/>/ {/>/ !{'p''}} >> $line.fasta
        done
    done
```

Sequences from genomic clusters and MIBiG were combine

```python
%%bash
cat TF_domains_v2 | while read line
    do
    echo $line
    cat 'TF_clusteric/'$line'_all.fasta_clusteric.id.fasta' 'MIBIG/'$line'_all.fasta' > TF_clusteric/$line'_clusteric.fasta'
    done
```

## Phylogenetic Trees

Alignments of genomic and clusteric TF files were done with muscle

```python
%%bash
ls TF_genomic/*genomic.fasta | while read line
    do
    muscle -in $line -out $line.afa -maxiters 1 -diags1 -sv
    mv $line.afa alignments/all/
    done
```

```python
%%bash
ls TF_clusteric/*clusteric.fasta | while read line
    do
    muscle -in $line -out $line.afa -maxiters 1 -diags1 -sv
    mv $line.afa alignments/clusteric/
    done
```

Phylogenetic trees were calulated with FastTree of genomic and clusteric sequences

```python
%%bash
ls alignments/all/*afa | while read line
    do
    FastTree < $line > $line.nwk
    mv $line.nwk trees/all/
    done
```

```python
%%bash
ls alignments/clusteric/*afa | while read line
    do
    FastTree < $line > $line.nwk
    mv $line.nwk trees/clusteric/
    done
```
