## Search of CA domains in Cuatro Cienegas and Abyssal genomes

Actual accesion number of CA domains were recovered from Pfam-A.hmm file using CAdomain.txt file

```python
%%bash
cut CAdomain.txt -f1 | while read line; do grep $line ../GENOMES/Pfam-A.hmm | cut -d ' ' -f 4; done > CAdomain_v2
```

Each Pfam profile in CAdomain_v2 were recovered from file Pfam-A.hmm and used to look for in each genome with function hmmsearch

```python
%%bash
ls sequences/*faa | while read line; do cat CAdomain_v2 | hmmfetch -f ../GENOMES/Pfam-A.hmm - |hmmsearch - $line > $line.out; done
```

To recover the protein sequence in each genome script PFAM-parsel.pl were used

```python
%%bash
ls sequences/*.out | while read line; do perl ../../Sources/pFAM-parser.pl $line; done
```

Aminoacid sequence were recovered

```python
%%bash
ls sequences/*out-PF* | while read list; do cut $list -f2- | tail -n +2 | while read id; do echo '>'$id >> $list.fasta; grep -A 1000 -w -h $id sequences/*.faa | sed -n -e '1,/>/ {/>/ !{'p''}} >>$list.fasta; done; done
```

Sequences of same protein family were grouped in a same file

```python
%%bash
cat CAdomain_v2 | while read line ; do echo $line ; cat sequences/*$line*fasta > sequences/$line'_all.fasta'; done
```

All CA-proteins IDs were combined in a same file to delete duplicated ID

```python
%%bash
grep 'fig' sequences/PF* -h | tr -d '>' | sort | uniq > sequences/CA_genomes.id
```

CA-proteins sequences were recovered using MIBIG/MIBiG_CA.id file

```python
%%bash
cat sequences/CA_genomes.id | while read id
    do
    echo '>'$id >> sequences/CA_genomes.fasta
    grep -A 1000 -w -h $id sequences/*faa | sed -n -e '1,/>/ {/>/ !{'p''}} >> sequences/CA_genomes.fasta
done
```

## Search of CA domains in MIBiG clusters

Proteins with CA domain in MIBiG clusters were search

```python
%%bash
ls MIBiG/*fasta |while read line; do cat CAdomain_v2 | hmmfetch -f ../GENOMES/Pfam-A.hmm - |hmmsearch - $line > $line.out; done
```

ID of proteins were recovered with script pFAM-parser.pl

```python
%%bash
ls MIBiG/*out | while read line; do perl ../../Sources/pFAM-parser.pl $line; done
```

IDs were corrected to match with actual IDs

```python
%%bash
perl -p -i -e 's/\n/_/g' MIBiG/*PF*
```

```python
%%bash
perl -p -i -e 's/_BGC0/\nBGC0/g' MIBiG/*PF*
```

Aminoacid sequences were recovered

```python
%%bash
ls MIBiG/*out-PF* | while read list
    do
    tail $list -n +2 | while read id
        do
        echo '>'$id >> $list.fasta
        grep -A 1000 -w -h $id MIBiG/mibig_prot_seqs_2.0.fasta | sed -n -e '1,/>/ {/>/ !{'p''}} >> $list.fasta
    done
done
```

Sequences of same protein family were grouped in a same file

```python
%%bash
cat CAdomain_v2 | while read line ; do echo $line ; cat MIBiG/*$line*fasta > MIBiG/$line'_all.fasta'; done
```

All CA-proteins IDs were combined in a same file to delete duplicated ID

```python
%%bash
grep 'BGC' MIBiG/PF* -h | tr -d '>' | sort | uniq > MIBiG/MIBiG_CA.id
```

CA-proteins sequences were recovered using MIBIG/MIBiG_CA.id file

```python
%%bash
cat MIBiG/MIBiG_CA.id | while read id
    do
    echo '>'$id >> MIBiG/CA_MIBiG.fasta
    grep -A 1000 -w -h $id MIBiG/mibig_prot_seqs_2.0.fasta | sed -n -e '1,/>/ {/>/ !{'p''}} >> MIBiG/CA_MIBiG.fasta
done
```

## Histidine Kinase match

CA-proteins of MIBIG and genomic sequences were combine in one file

```python
%%bash
cat sequences/CA_genomes.fasta MIBiG/CA_MIBiG.fasta > HisKA/CA_all.fasta
```

Actual Pfam number were recovered from Pfam-A.hmm file

```python
%%bash
cut His-KA-Pfam.txt -f1 | while read line; do grep $line ../GENOMES/Pfam-A.hmm | cut -d ' ' -f 4; done > His-KA-Pfam_v2
```

Histidine Kinase phosphoacceptor domain were detected in CA domains

```python
%%bash
ls HisKA/*fasta | while read line; do cat His-KA-Pfam_v2 | hmmfetch -f ../GENOMES/Pfam-A.hmm - | hmmsearch - $line > $line.out; done
```

Protein IDs were recovered with pFAM-parser script

```python
%%bash
ls HisKA/*out | while read line; do perl ../../Sources/pFAM-parser.pl $line; done
```

MIBiG ID were corrected to match actual IDs

```python
%%bash
perl -p -i -e 's/\n/_/g' HisKA/*PF*
```

```python
%%bash
perl -p -i -e 's/_BGC0/\nBGC0/g' HisKA/*PF*
```

```python
%%bash
perl -p -i -e 's/_fig/\nfig/g' HisKA/*PF*
```

Aminoacid sequences were recovered 

```python
%%bash
ls HisKA/*out-PF* | while read list
    do
    tail $list -n +2 | while read id
        do
        echo '>'$id >> $list.fasta
        grep -A 1000 -w -h $id HisKA/CA_all.fasta | sed -n -e '1,/>/ {/>/ !{'p''}} >> $list.fasta
    done
done
```
