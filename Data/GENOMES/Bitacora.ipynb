# Buscando dominios selectos en genomas de Cuatro Cienegas

Con la funcion grep se buscaron los dominios de interes del archivo Dominios en el archivo Pfam-A.hmm

```python
%%bash 
cat Dominios | while read line; do grep $line Pfam-A.hmm; done > dominios_new
```

```python
%%bash
head dominios_new
```

Se elimino el texto "ACC " de cada linea con la funcion sustitute perl

```python
%%bash
perl -p -i -e 's/ACC\s*//' dominios_new
```

```python
%%bash
head dominios_new
```

Cada dominio se recupero su perfil HMM del archivo Pfam-A.hmm con la funcion hmmfetch y con cada perfil se busco dentro de los genomas con la funcion hmmsearch

```python
%%bash
ls genomas/*faa|while read line; do cat dominios_new | hmmfetch -f Pfam-A.hmm - |hmmsearch - $line > $line.out; done
```

Para recuperar la secuencia de proteina en cada genoma se utilizo el script PFAM-parser.pl 

```python
%%bash
ls genomas/*.out | while read line; do perl ../../Sources/pFAM-parser.pl $line; done
```

Para cada archivo se recupero solo su identificador y la secuencia de aminoacidos

```python
ls genomas/*out-PF?????.?? | while read list; do cut $list -f2- | tail -n +2 | while read id; do echo '>'$id >> $list.fasta; grep -A 1000 -w -h $id genomas/*.faa | sed -n -e '1,/>/ {/>/ !{'p''}} >>$list.fasta; done; done
```

Se concentraron las secuencias de cada genoma que correspondia al mismo dominio en un mismo archivo por dominio

```python
%%bash
cat dominios_new | while read line ; do  echo $line ; cat genomas/*$line*fasta > genomas/$line'_all.fasta'; done
```

# Buscando dominios selectos en genomas de MiBIG

Para las secuencias de MiBIG se realizo la busqueda de secuencias que corresponden a los dominios seleccionados

```python
%%bash
ls MiBIG/*faa |while read line; do cat dominios_new | hmmfetch -f Pfam-A.hmm - |hmmsearch - $line > $line.out; done
```

Para cada secuencia se recuperaros los id de las proteinas con el scrip pFAM-parser.pl

```python
%%bash
ls MiBIG/*.out | while read line; do perl ../../Sources/pFAM-parser.pl $line; done
```

Ya que los archivos resultantes tienen su identificador de cluster y de proteina en lineas separadas, primero se sustituyo todos los saltos de lineas por tabulacions

```python
%%bash
perl -p -i -e 's/\n/\t/g' MiBIG/*PF*
```

Luego se inserto un salto de linea antes de cada identificar de cluster

```python
%%bash
perl -p -i -e 's/BGC0/\nBGC0/g' MiBIG/*PF*
```

Despues todas las tabulaciones se sustituyeron por "_"

```python
%%bash
perl -p -i -e 's/\t/_/g' MiBIG/*PF*
```

Finalmente, los "_" al final de cada identificador se elimino

```python
%%bash
perl -p -i -e 's/_\n/\n/' MiBIG/*PF*
```

Se recuperaron las secuencias de aminoacidos de los identificadores

```python
%%bash
ls MiBIG/*out-PF?????.?? | while read list; do cut $list -f2- | tail -n +2 | while read id; do echo '>'$id >> $list.fasta; grep -A 1000 -w -h $id MiBIG/*.faa | sed -n -e '1,/>/ {/>/ !{'p''}} >>$list.fasta; done; done
```

```python
%%bash
cat dominios_new | while read line; do echo $line ; cat MiBIG/*$line.fasta genomas/$line'_all.fasta' >> fasta_sequences/$line.fasta; done
```

Se realizo el alineamiento de las secuencias por familia

```python
%%bash
cat dominios_new | while read line; do muscle -in fasta_sequences/$line.fasta -out alignments/$line.afa -maxiters 1 -diags1 -sv; done
```

Con los archivos de alineamientos se calcularon los arboles filogeneticos con FastTree

```python
%%bash
cat dominios_new | while read line; do FastTree < alignments/$line.afa > trees/$line.nwk; done
```

# Adicion de metadata

Se extrajeron los metadatos de cada genoma analizado por antiSMASH con el scrip id_metadata.sh

```python
%%bash
ls -d antiSMASH_analisis/*/ | while read folder; do bash id_metadata.sh $folder ; done
```

Todos los metadatos se concentraron en un solo archivo

```python
%%bash
ls antiSMASH_analisis/*/metadata.txt | while read data; do cat $data >> metadata_all.txt; done
```

Los metadatos de genomas de MiBIG se reacomodaron del archivo MiBIG_DB.faa.meta

```python
%%bash
paste <(cut -f 1 MiBIG/MiBIG_DB.faa.meta) <(cut -f 4 MiBIG/MiBIG_DB.faa.meta) <(cut -f 2 MiBIG/MiBIG_DB.faa.meta) <(cut -f 3 MiBIG/MiBIG_DB.faa.meta) >> metadata_MiBIG.meta
```

Se recuperaron los metadatos de las secuencias de proteina usadas para los arboles filogeneticos

```python
%%bash
ls fasta_sequences/ | while read file
do
    list_CC=$(grep 'fig' fasta_sequences/$file | cut -d '>' -f2)
    list_BGC=$(grep 'BGC' fasta_sequences/$file | cut -d '>' -f2)
    echo -e 'ID''\t''Origin__autocolor''\t''Clusteric__autocolor''\t''Organism__autocolor''\t''Product__autocolor''\t''Type__autocolor' >> metadata/$file'_meta.tsv'
    for line in $list_CC
    do
        result=$(grep -r $line metadata_all.txt)
        if [ -z $result];
        then
            organism=$(echo $line | cut -d '|' -f2 | cut -d '.' -f1-2 )
            org=$(grep $organism organism.tsv | cut -f2)
            echo -e $line'\t'Cuatro_cienegas'\t'Extraclusteric'\t'$org'\t''\t' >> metadata/$file'_meta.tsv'
            else
            echo -e $line'\t'Cuatro_cienegas'\t'Intraclusteric'\t'$(grep $line metadata_all.txt | cut -f2)'\t'$(grep $line metadata_all.txt | cut -f3)'\t'$(grep $line metadata_all.txt | cut -f4) >> metadata/$file'_meta.tsv'
            fi
        done
    for line in $list_BGC
    do
        line2=$(echo $line | cut -d '_' -f1)
        echo -e $line'\t'MiBIG'\t'Intraclusteric'\t'$(grep $line2 metadata_MiBIG.meta | cut -f2)'\t'$(grep $line2 metadata_MiBIG.meta | cut -f3)'\t'$(grep $line2 metadata_MiBIG.meta | cut -f4) >> metadata/$file'_meta.tsv'
        done
    done
```

Para contruir los arboles solo de secuencias dentro de clusteres primero se recuperaros los metadatos solo de secuencias de clusteres

```python
%%bash
ls fasta_sequences/ | while read file
    do
     echo -e 'ID''\t''Region__autocolor''\t''Organism__autocolor''\t''Type_1__autocolor''\t''Type_2__autocolor''\t''Product__autocolor''\t''Similarity' >> metadata/intraclusteric/$file'_metadata.tsv'
    grep '>fig' fasta_sequences/$file | cut -d '>' -f 2 | while read line
        do
        grep -w $line metadata/intraclusteric/all_metadata.tsv >> metadata/intraclusteric/$file'_metadata.tsv'
        done
    grep '>BGC' fasta_sequences/$file | cut -d '>' -f 2  | while read line
        do
        clu=$(echo $line | cut -d '_' -f 1)
        echo -e $line'\t'$(grep -w $clu metadata/intraclusteric/all_metadata.tsv | cut -f2)'\t'$(grep -w $clu metadata/intraclusteric/all_metadata.tsv | cut -f3)'\t'$(grep -w $clu metadata/intraclusteric/all_metadata.tsv | cut -f4)'\t'$(grep -w $clu metadata/intraclusteric/all_metadata.tsv | cut -f5)'\t'$(grep -w $clu metadata/intraclusteric/all_metadata.tsv | cut -f6)'\t'$(grep -w $clu metadata/intraclusteric/all_metadata.tsv | cut -f7)'\t'$(grep -w $clu metadata/intraclusteric/all_metadata.tsv | cut -f8) >> metadata/intraclusteric/$file'_metadata.tsv'
        done
    done
```

Los metadatos se usaron para recuperar las secuencias fasta solo de id dentro de clusters

```python
%%bash
ls metadata/intraclusteric/PF* | while read list
    do
    cut $list -f1 | tail -n +2 | grep 'fig' - | while read id
        do
        name=$(echo $list | cut -d '/' -f 3 | cut -d '.' -f1-2)
        echo '>'$id  >> fasta_sequences/intraclusteric/$name'.fasta'
        grep -A 1000 -w -h $id genomas/*.faa | sed -n -e '1,/>/ {/>/ !{'p''}} >> fasta_sequences/intraclusteric/$name'.fasta'
        done
    done
```

```python
%%bash
ls metadata/intraclusteric/PF* | while read list
    do
    cut $list -f1 | tail -n +2 | grep 'BGC' - | while read id
        do
        name=$(echo $list | cut -d '/' -f 3 | cut -d '.' -f1-2)
        echo '>'$id  >> fasta_sequences/intraclusteric/$name'.fasta'
        grep -A 1000 -w -h $id MiBIG/*.faa | sed -n -e '1,/>/ {/>/ !{'p''}} >> fasta_sequences/intraclusteric/$name'.fasta'
        done
    done%%bash
```

Se realizo el alineamiento de las secuencias por familia

```python
%%bash
cat dominios_new | while read line
    do muscle -in fasta_sequences/intraclusteric/$line.fasta -out alignments/intraclusteric/$line.afa -maxiters 1 -diags1 -sv; done
```

Con los archivos de alineamientos se calcularon los arboles filogeneticos con FastTree

```python
%%bash
cat dominios_new | while read line; do FastTree < alignments/intraclusteric/$line.afa > trees/intraclusteric/$line.nwk; done
```

# Arbol filogeneticos SARP

Se construyo un arbol filogenetico para proteinas con estructura de Streptomyces Antibiotic Protein (SARP) PF00486 + PF03704

De las secuencias con dominio PF00486 se buscaron las proteinas con dominio PF03704

```python
%%bash
hmmfetch Pfam-A.hmm PF03704.19 | hmmsearch - fasta_sequences/intraclusteric/PF00486.30.fasta > SARPs/PF00486+PF03704.out
```

Se recuperaron los identificadores de las secuencias con el script

```python
%%bash
perl ../../Sources/pFAM-parser.pl SARPs/PF00486+PF03704.out
```

Con los identificadores se buscaron las secuencias fasta del archivo original

```python
%%bash
tail -n +2 SARPs/PF00486+PF03704.out-PF03704.19 | while read id
    do
    echo '>'$id >> SARPs/PF00486+PF03704.fasta
    grep -A 1000 -w -h $id fasta_sequences/intraclusteric/PF00486.30.fasta | sed -n -e '1,/>/ {/>/ !{'p''}} >> SARPs/PF00486+PF03704.fasta
    done
```

Se realizo el alineamiento de las secuencias

```python
%%bash
muscle -in SARPs/PF00486+PF03704.fasta -out SARPs/PF00486+PF03704.afa -maxiters 1 -diags1 -sv
```

Se calculo el arbol filogenetico con los alineamientos

```python
%%bash
FastTree < SARPs/PF00486+PF03704.afa > SARPs/PF00486+PF03704.nwk
```

Se recuperaron los metadatos de las secuencias fasta

```python
%%bash
echo -e 'ID''\t''Region__autocolor''\t''Organism__autocolor''\t''Type_1__autocolor''\t''Type_2__autocolor''\t''Product__autocolor''\t''Similarity' >> SARPs/PF00486+PF03704_metadata.tsv
grep '>fig' SARPs/PF00486+PF03704.fasta | cut -d '>' -f 2 | while read line
    do
    grep -w $line metadata/intraclusteric/all_metadata.tsv >> SARPs/PF00486+PF03704_metadata.tsv
    done
grep '>BGC' SARPs/PF00486+PF03704.fasta | cut -d '>' -f 2  | while read line
    do
    clu=$(echo $line | cut -d '_' -f 1)
    echo -e $line'\t'$(grep -w $clu metadata/intraclusteric/all_metadata.tsv | cut -f2)'\t'$(grep -w $clu metadata/intraclusteric/all_metadata.tsv | cut -f3)'\t'$(grep -w $clu metadata/intraclusteric/all_metadata.tsv | cut -f4)'\t'$(grep -w $clu metadata/intraclusteric/all_metadata.tsv | cut -f5)'\t'$(grep -w $clu metadata/intraclusteric/all_metadata.tsv | cut -f6)'\t'$(grep -w $clu metadata/intraclusteric/all_metadata.tsv | cut -f7)'\t'$(grep -w $clu metadata/intraclusteric/all_metadata.tsv | cut -f8) >> SARPs/PF00486+PF03704_metadata.tsv
    done
```
